```python
#  Divide el conjunto de datos en un conjunto de entrenamiento y uno de prueba, utilizando los métodos 
#    suministrados por sckitlearn (train_test_split).
     # El conjunto de prueba debe tiene el 30% del total de los datos. 
    
from sklearn.model_selection import train_test_split

x=df[['radius','texture','perimeter','area','smoothness','compactness','concavity','concave','symmetry','fractal']]
y=df['class']

X_entrenamiento, X_prueba, y_entrenamiento, y_prueba = train_test_split(x,y,random_state=3, test_size = 0.3)
```

```python
# Determina el tamaño del conjunto de entrenamiento y el del conjunto de pruebas obtenido e 
    #identifica en cada conjunto, cuantos datos de cada clase existen. Imprim3 estos valores.
    
print("Tamaño del conjunto de entrenamiento: {}".format(X_entrenamiento.shape))
print("Tamaño del conjunto de entrenamiento: {}".format(y_entrenamiento.shape))
print("Tamaño del conjunto de pruebas: {}".format(X_prueba.shape))
print("Tamaño del conjunto de pruebas: {}".format(y_prueba.shape))

print("\nCantidad de datos de entrenamiento de clase Maligna")
print(np.sum(y_entrenamiento=='M'))
print("Cantidad de datos de entrenamiento de clase Benigna")
print(np.sum(y_entrenamiento=='B'))
print("Cantidad de datos de prueba de clase Maligna")
print(np.sum(y_prueba=='M'))
print("Cantidad de datos de prueba de clase Benigna")
print(np.sum(y_prueba=='B'))

```

```python
# Comprueba si las proporción de clases en cada conjunto (prueba y entrenamiento) es similar
print(np.sum(y_entrenamiento=='M'))
print(np.sum(y_entrenamiento=='B'))

print(np.sum(y_prueba=='M'))
print(np.sum(y_prueba=='B'))
```

```python
# Crea un clasificador de tipo vecino mas cercano (knn), mediante KNeighborsClassifier, entrena y prueba el exito en la clasificación
#    mediante los conjuntos de prueba y entrenamiento.

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_entrenamiento, y_entrenamiento)

y_pred = knn.predict(X_prueba)
print("\nPredicciones conjunto de prueba:\n{}".format(y_pred))
print("Resultado de la prueba:",knn.score(X_prueba, y_prueba))

```

```python
# Imprime la matriz de confusión de acuerdo a los resultados obtenidos.
from sklearn.metrics import confusion_matrix
print("\nMatriz de confusión:")
skcm = confusion_matrix(y_prueba, y_pred)

# colocar en un dataframe para imprimir las etiquetas
skcm = pd.DataFrame(skcm, columns=['Predice: Maligno','Predice: Benigno'])
skcm['actual'] = ['Maligno','Benigno']
skcm = skcm.set_index('actual')
print(skcm)
```

```python
# Calcula a partir de la matriz de confusión las métricas de rendimiento: Precisión, Recuperación y F1_Score. Utiliza classification_report
from sklearn.metrics import classification_report

print("\nReporte de la Clasificación:") 
print(classification_report(y_prueba, y_pred, target_names=['Maligno', 'Benigno']))
```

```python
# Análisis y descripción de los resultados obtenidos en las métricas.

# Para la clase de tumores malignos la precisión es del 91% y para la clase tumores benignos es del 87%
# Se encuentra más precisión en la clasificación de la clase maligna,pese a esto se puede decir que
# esta clasificación tiene una alta precisión. 
# Por otra parte el recall para los tumores malignos es del 93% y para los tumores benignos del 84%, 
# Lo cual nos indica que algunos tumores se predijeron incorrectamente como benignos cuando realmente eran malignos.
# En cuanto al f1-score, para el tumor maligno es del 92% y para el tumor benigno es del 85%.
# En manera de conclusión, pese al porcentaje de error en la predicción que no es tan alto, 
# se puede decir que es un modelo eficiente para la clasificación de tumores.
```

```python
# Realiza validación cruzada para verificar la validez de los resultados con diferentes conjuntos de prueba y entrenamiento.
#     Utiliza el objeto cross_val_score y divide en 10 conjuntos.

from sklearn.model_selection import cross_val_score
scores = cross_val_score(knn, x, y, cv=10)
```

```python
#  Muestra el score obtenido en cada iteración ( porcentaje de acierto )
scores
```

```python
# calcula el promedio de acierto en la clasificación y la desviación estandar. 
#     Tener en cuenta : print("Precisión: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))

print("Promedio acierto: {}".format(scores.mean()))
print("Desviación estándar: {}".format(scores.std()))

print("Precisión: %0.2f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
